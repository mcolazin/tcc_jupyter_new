## Caderno 1 - Agregação dos Dados

```r
library(tidyverse)
```

#### Carregamento dos Dados

```r
# Dados do Enade:
enade_data = read.csv("data/01a_enade_2019_summary_mean_grade_by_citie.csv", encoding = "UTF-8")
enade_data = enade_data %>%
  rename("cod_municipio_ibge" = 1)

# Dados da Anatel sumarizados por município:
anatel_data_muni = read.csv("data/01c_anatel_internet_access_processed.csv", encoding = "UTF-8") %>% select(-X)
# Dados da Anatel sumarizados por UF:
anatel_data_uf = read.csv("data/01d_anatel_internet_access_processed_by_uf.csv", encoding = "UTF-8")

# Dados de remuneração:
pnad_wage_data_uf = read.csv("data/01b_pnad_mean_sd_uf.csv", encoding = "UTF-8") %>% select(-X)

# Dados do índice de desigualdade - Gini
gini_data_uf = read.csv("data/01_gini_processed.csv", encoding = "UTF-8", sep=";")

# Dados do PIB
pib_data_muni = read.csv("data/02_pib_processed.csv")

# Dados de IDH
idh_data_muni = read.csv("data/03_idh_processed_mean_uf_2019.csv")
```

### Unificação dos dados por UF e código IBGE do município

```r
# Enade + Anatel por Municípios
enade_anatel_munic_merge = merge(enade_data, anatel_data_muni, by.x = "cod_municipio_ibge", by.y = "cod_municipio_ibge") %>% select(-municipio)
```

```r
# Enade + Anatel Muni + pnad_remuneracao_uf
enade_anatelMun_remuner_merge = merge(enade_anatel_munic_merge, pnad_wage_data_uf, by.x = "UF", by.y = "UF")
```

```r
# Enade + Anatel_mun + remuner_uf + gini_uf
enade_anatelMun_remuner_gini_merge = merge(enade_anatelMun_remuner_merge, gini_data_uf, by.x = "UF", by.y = "uf")
```

```r
# Enade + Anatel_mun + remuner_uf + gini_pib
enade_anatelMun_remuner_gini_pib_merge = merge(enade_anatelMun_remuner_gini_merge, pib_data_muni[c("cod_mun", "pib", "pib_p_capta")], by.x = "cod_municipio_ibge", by.y = "cod_mun" )
```

```r
# Enade + Anatel_mun + remuner_uf + gini_pib + idh
enade_anatelMun_remuner_gini_pib_idh_merge = merge(enade_anatelMun_remuner_gini_pib_merge, idh_data_muni, by.x = "UF", by.y = "uf")
```

```r
enade_anatelMun_remuner_gini_pib_idh_merge
```

### Gravação dos Dados Integrados para CSV

```r
write.csv(enade_anatelMun_remuner_gini_pib_idh_merge, row.names=FALSE, "integrated_processed_enem_anatel_ibge_2019.csv")
```
